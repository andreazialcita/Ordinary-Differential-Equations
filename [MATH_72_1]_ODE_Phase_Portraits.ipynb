## 3D Phase Portraits

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp


r = 0.3179
K = 1e10
a = 3.5e-6
b = 5.434e-9
k1 = 6.543e-11
k2 = 1.773e-11
c1 = 0.2743
c2 = 0.1

A = r / c1
B = a * K * k1 / c1
C = b * K * k2 / c1
D = K * k1 / c1
E = K * k2 / c1
F = c2 / c1

initial_conditions = [
    [0.1, 0.1, 0.1],
    [0.5, 0.5, 0.2],
    [0.8, 0.2, 0.3],
    [1.0, 0.5, 0.5]
]

t_span = (0, 50)
t_eval = np.linspace(*t_span, 500)

def system(t, state):
    x, y, z = state
    dxdt = A * x * (1 - x) - B * x * y + C * x * z
    dydt = D * x * y - y
    dzdt = E * x * z - F * z
    return [dxdt, dydt, dzdt]

# Plot 3D phase trajectories
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

for ic in initial_conditions:
    sol = solve_ivp(system, t_span, ic, t_eval=t_eval)
    ax.plot(sol.y[0], sol.y[1], sol.y[2], lw=2)

ax.set_title('3D Phase Trajectories')
ax.set_xlabel('x (Tumor cells)')
ax.set_ylabel('y (NK cells)')
ax.set_zlabel('z (TAMs)')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the system
def system(t, vars, A, B, C, D, E, F):
    x, y, z = vars
    dxdt = A * x * (1 - x) - B * x * y + C * x * z
    dydt = D * x * y - y
    dzdt = E * x * z - F * z
    return [dxdt, dydt, dzdt]

# Parameters for 56ug
r = 0.35
K = 5e9
a = 3.5e-6
b = 2e-6
k2 = 0.8e-11
c2 = 0.1
T = 1e9
N0 = 1e4
M0 = 1e5

k1_56ug = 2.34e-10  # proliferation rate of NK
c1_56ug = 0.1355    # apoptotic rate of NK

A = r / c1_56ug
B = a * N0 / c1_56ug
C = b * M0 / c1_56ug
D_56ug = K * k1_56ug / c1_56ug
E = K * k2 / c1_56ug
F = c2 / c1_56ug

params_56ug = dict(A=A, B=B, C=C, D=D_56ug, E=E, F=F)

# Initial conditions as in the reference figure
initial_conditions = [
    [0.8, 0.95, 0.95],
    [0.6, 0.85, 0.80],
    [0.3, 0.80, 0.70],
    [0.1, 0.50, 0.50],
    [0.08, 0.20, 0.15],
    [0.2, 1.33, 0.0]
]

# Time span
t_span = (0, 15)
t_eval = np.linspace(*t_span, 500)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for ic in initial_conditions:
    sol = solve_ivp(system, t_span, ic, args=tuple(params_56ug.values()), t_eval=t_eval)
    ax.plot(sol.y[0], sol.y[1], sol.y[2], lw=2)
    ax.text(sol.y[0][0], sol.y[1][0], sol.y[2][0], f"({ic[0]:.2f},{ic[1]:.2f},{ic[2]:.2f})", fontsize=8)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(r'(a) $D > 1$')

# Plot the fixed point in 3D
x_fp = 1 / D_56ug
y_fp = A / B * (1 - 1 / D_56ug)
z_fp = 0
ax.scatter(x_fp, y_fp, z_fp, color='black', s=50, label='Fixed Point')
print(x_fp, y_fp, z_fp)

ax.legend()

plt.tight_layout()
plt.show()
```
